```python
election_data=new_data.fillna(0)
```

## Analyzation Between PCname and State

```python
grouped=election_data['PC Name'].groupby(election_data['State']).count()
```

```python
grouped
```

## Diagramatic Representation Of Above Analysis

```python
mylabels=['Andaman&Nicobar','Andhra','Arunachal','Assam','Bihar','Chandigarh','Chhattisgarh','Dadra&Nagar Haveli','Daman&Diu','Delhi','Goa','Gujarat','Haryana','Himachal','Jammu&Kashmir','Jharkhand','Karnataka','Kerala','Lakshadweep','MadhyaPradesh','Maharashtra','Manipur','Meghalaya','Mizoram','Nagaland','Orissa','Puducherry','Punjab','Rajasthan','Sikkim','TamilNadu','Telangana','Tripura','UttarPradesh', 'Uttarakhand','West Bengal']
plt.figure(figsize=(20,40))
plt.pie(grouped,labels=mylabels,autopct='%1.1f%%')
plt.legend()
plt.show()
```

**From the above analysis and visualization,UttarPradesh has the high number of PC.**

## Analyzation Between State and Votes

```python
vote=election_data['Votes'].groupby(election_data['State']).sum()
```

```python
vote
```

## Diagramatic Representation Of Above Analysis

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(vote)
```

**From above analysis we can understand that most number of votes were submited in UttarPradesh**

## Analyzation Between Party And Votes

```python
party_vote=election_data['Votes'].groupby(election_data['Party']).sum()
```

```python
party_vote
```

## Diagramatic Representaion Of Above Analysis

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.grid()
sns.lineplot(party_vote)
```

**From above analysis Bharatiya Janta Party(BJP) got most number of votes**
